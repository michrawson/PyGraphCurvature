# Import libraries and define function to make graphs.

```python
import networkx as nx
from example_graphs import make_chung_lu_clustering_graph
from compare_curvatures import compare_curvatures, curv_vs_time
from graph_curvature import forman_curvature, forman_aug_curvature, lly_curvature
import numpy as np
import matplotlib.pyplot as plt

def get_graph(graph_type,gamma=0.):
    if graph_type == 'chung_lu':
        n = 1000
        alpha = 1.
        beta = .1
        gamma = .0
        G = make_chung_lu_clustering_graph(n,alpha,beta,gamma)

    elif graph_type == 'erdos':
        n = 1000
        G = nx.erdos_renyi_graph(n,p=.003)
        
    else:
        raise NotImplementedError()                                    

    return G
```

# Create Erdos graph, plot adjacency matrix, compute curvatures, and 
# plot their histograms. 

```python
graph_type = 'erdos'
G = get_graph(graph_type)
G_fcurv, G_fcurv_node = forman_curvature(G)
G_fcurv_aug, G_fcurv_aug_node = forman_aug_curvature(G)
LLY_curv, LLY_curv_node = lly_curvature(G)
compare_curvatures(G, G_fcurv, G_fcurv_aug, LLY_curv)
```

# Create Chung-Lu graph, plot adjacency matrix, compute curvatures, and 
# plot their histograms. 

```python
graph_type = 'chung_lu'
G = get_graph(graph_type)
G_fcurv, G_fcurv_node = forman_curvature(G)
G_fcurv_aug, G_fcurv_aug_node = forman_aug_curvature(G)
LLY_curv, LLY_curv_node = lly_curvature(G)
compare_curvatures(G, G_fcurv, G_fcurv_aug, LLY_curv)
```
