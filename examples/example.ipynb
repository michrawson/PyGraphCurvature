```python

import networkx as nx
from example_graphs import make_chung_lu_clustering_graph
from compare_curvatures import compare_curvatures, curv_vs_time
from graph_curvature import forman_curvature, forman_aug_curvature, lly_curvature_approx, lly_curvature
import numpy as np
import matplotlib.pyplot as plt

def get_graph(graph_type,gamma=0.):
    if graph_type == 'chung_lu':
        n = 1000
        alpha = 1.
        beta = .1
        gamma = .0
        G = make_chung_lu_clustering_graph(n,alpha,beta,gamma)

    elif graph_type == 'erdos':
        n = 1000
        G = nx.erdos_renyi_graph(n,p=.003)
        
    else:
        raise NotImplementedError()                                    

    return G
```

```python
# Parameters

alpha = 0.9 
```

```python
graph_type = 'erdos'
G = get_graph(graph_type)
G_fcurv = forman_curvature(G)
G_fcurv_aug = forman_aug_curvature(G)
LLY_curvatures = lly_curvature(G, alpha)
compare_curvatures(G, G_fcurv, G_fcurv_aug, LLY_curvatures)
```
